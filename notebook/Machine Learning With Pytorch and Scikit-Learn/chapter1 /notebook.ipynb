# Giving Computers the Ability to Learn from Data

## 机器学习

提供一种将数据转换为知识的方法。

### 分类

- 监督学习
  - 分类任务：通过已知的输入输出数据集来预测结果，也可称为标签学习
  - 回归：通过数据预测持续的输出
- 非监督学习(探索发现隐藏结构，提取有意义的数据)
  - 聚类：将数据划分为不同的组别
  - 降维数据压缩，降低噪音
- 强化学习
  - 奖励信号

### 机器学习路线图

- 数据收集
- 数据清理
  - 缺失数据处理
  - 初始特征提取和选择
- 生成训练数据集
- 处理训练数据集
  - 特征缩放
  - 维度减少
    - 特征选择
    - 特征提取
- 模型选择
- 模型训练
